## Импортируем библиотеку sqlite3

```python
import sqlite3 as sql
```

### Упражнение 1
Создайте базу данных (БД) и курсор к ней, чтобы далее работать с этой БД.

```python
bd = sql.connect('Money.db')
cur = bd.cursor()
```

### Упражнение 2
Создайте в новой БД таблицу с пятью столбцами. Не забывайте сохранять выполненные изменения в БД с помощью метода commit().

```python
cur.execute('''CREATE TABLE IF NOT EXISTS cash (money_id INTEGER NOT NULL PRIMARY KEY,
            valute TEXT NOT NULL, country TEXT NOT NULL, symbol TEXT NOT NULL, in_rubles INTEGER NOT NULL);
            ''')
```

### Упражнение 3
Добавьте пять записей в БД.

```python
cur.execute('''INSERT INTO cash (money_id, valute, country, symbol, in_rubles)
            VALUES (1, "USD", "USA", "$", 89)
            ''')

cur.execute('''INSERT INTO cash (money_id, valute, country, symbol, in_rubles)
            VALUES (2, "EUR", "Grece", "€", 96)
            ''')

cur.execute('''INSERT INTO cash (money_id, valute, country, symbol, in_rubles)
            VALUES (3, "TRC", "Turkie", "₺", 3.1)
            ''')

cur.execute('''INSERT INTO cash (money_id, valute, country, symbol, in_rubles)
            VALUES (4, "CNY", "China", "¥", 12.5)
            ''')

cur.execute('''INSERT INTO cash (money_id, valute, country, symbol, in_rubles)
            VALUES (5, "KZT", "Kazakhstan", "₸", 0.19)
            ''')

bd.commit()
```

### Упражнение 4
Запросите 3 записи из БД.

```python
cur.execute('SELECT money_id, valute, country, symbol, in_rubles FROM cash WHERE money_id <= 3')
print(cur.fetchall())
```

### Упражнение 5
Удалите одну из записей вашей БД.

```python
cur.execute('DELETE FROM cash WHERE money_id = 2')
cur.execute('SELECT money_id, valute, country, symbol, in_rubles FROM cash WHERE money_id <= 3')
print(cur.fetchall())
bd.commit()
```
